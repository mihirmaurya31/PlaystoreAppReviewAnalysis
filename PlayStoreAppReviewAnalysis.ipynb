# Text Data Cleaning by Removal of Punctuation and Stopwords

```python
import nltk
from nltk.corpus import stopwords
import gensim
from gensim.utils import simple_preprocess
```

```python
# download stopwords
nltk.download('stopwords')
stopwords.words('english')
```

```python
# Simple_preprocess converts a string into a series of lowered case tokens
# For example
gensim.utils.simple_preprocess(play_store_df['Translated_Review'][0])
```

Removal of punctuation

```python
import string
string.punctuation
```

```python
def remove_punctuation(text):
  text_punc_remove = [word   for word in text   if word not in string.punctuation]
  text_punc_join = ''.join(text_punc_remove)
  return text_punc_join
```

```python
play_store_df['text_punc_join'] = play_store_df['Translated_Review'].apply(remove_punctuation)
```

```python
play_store_df.head()
```

Removal of Stopwords and short words (less than 2 characters)

```python
def preprocess(text):
  result = []
  for token in gensim.utils.simple_preprocess(text):
    if token not in stopwords.words('english') and len(token) > 2:
      result.append(token)
  return result
```

```python
play_store_df['reviews_punc_stopwords_removed'] = play_store_df['text_punc_join'].apply(preprocess)
```

```python
play_store_df.head()
```

```python
play_store_df['reviews_punc_stopwords_removed_join'] = play_store_df['reviews_punc_stopwords_removed'].apply(lambda x: ' '.join(x))
```
